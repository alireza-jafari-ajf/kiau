**arange** **VS** **range**

```python
N = 1000000
my_arr = np.arange(N)
my_list = [i for i in range(N)]
```

**Numpy ==> arange**

```python
%time for _ in range(10): my_arr2 = 2*my_arr
```

**Python ==> range**

```python
%time for _ in range(10): my_list2 = [2*x for x in my_list]
```

**Pandas**

```python
import pandas as pd
```

```python
p = pd.read_csv("EXELS/Weather_Data_in_India_from_1901_to_2017.csv")
p
```

```python
df = pd.DataFrame(p)
df
```

**Matplotlib**

```python
from matplotlib import pyplot as plt 
```

**Data Visualization**

```python
fig = plt.figure(figsize=(6,5))
plt.plot(df["YEAR"] , df["DEC"] , color="green")
plt.plot(df["YEAR"] , df["JAN"] , color="red")
plt.title("Weather Of India Country")
plt.xlabel("Year")
plt.ylabel("Months")
plt.grid()
plt.legend()
plt.show()
```
